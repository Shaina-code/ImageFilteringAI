```python
#gender prediction
#demonstration 1

from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
img1=cv2.imread(r'C:\Users\shiva\Downloads\man image.jpeg')
plt.imshow(img1[:,:,::-1])
plt.show()
```

```python
result=DeepFace.analyze(img1,actions=['gender'])
```

```python
print("gender :", result)
```
